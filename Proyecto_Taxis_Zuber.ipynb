# Prueba de Hipotesis

### "La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos".

##### En este caso estamos comparando las medias de dos grupos independientes; viajes en sábados lluviosos vs. viajes en sábados no lluviosos de una variable numérica, por eso la prueba estadística adecuada es una prueba t de Student para medias de dos muestras numericas independientes. Esta prueba nos ayudará a poder determinar si la diferencia observada entre las medias de los dos grupos es estadísticamente significativa o no.

### Necesitamos seleccionar solo los viajes que cumplen ciertas condiciones:
#### La ubicación de inicio es el "Loop"
#### La fecha del viaje es un sábado.
#### Las condiciones climáticas: "lluvioso" y "no lluvioso" Good o Bad

```python
from scipy import stats as st

# Filtramos los dias por Sabado (Se filtro una parte anteriormente en SQL)
weather_df['start_ts_day_of_week'] = weather_df['start_ts'].dt.day_name()
saturday_trips = weather_df[weather_df['start_ts_day_of_week'] == 'Saturday'].copy()

# Separamos el sabado por condiciones climaticas
# Suponemos que Bad significa condicion adversa y Good buen clima
saturday_bad_weather = saturday_trips[saturday_trips['weather_conditions'] == 'Bad']['duration_seconds']
saturday_good_weather = saturday_trips[saturday_trips['weather_conditions'] == 'Good']['duration_seconds']

#Verificamos si hay datos significativos para la prueba
print(f"Número de viajes en sábados con clima 'Bad': {len(saturday_bad_weather)}")
print(f"Número de viajes en sábados con clima 'Good': {len(saturday_good_weather)}")

# Establecemos (alpha). 0.05 es el estandar comunmente utilizado
alpha = 0.05

# Prueba de T-test
# Usamos equal_var=False Porque no sabemos si las varianzas de ambos grupos son iguales
if len(saturday_bad_weather) > 1 and len(saturday_good_weather) > 1:
    results = st.ttest_ind(
        saturday_bad_weather,
        saturday_good_weather,
        equal_var=False)

    print('\nP-value:', results.pvalue)

    # Verificamos el valor con respecto a alpha
    if results.pvalue < alpha:
        print("Rechazamos la hipótesis nula: La duración promedio de los viajes cambia los sábados lluviosos (o con mal clima).")
    else:
        print("No rechazamos la hipótesis nula: No hay suficiente evidencia para decir que la duración promedio de los viajes cambia los sábados lluviosos (o con mal clima).")
else:
    print("\nNo hay suficientes datos en uno o ambos grupos para realizar la prueba t.")
```

#Conclusiones:

#### Tenemos evidencia estadística para concluir que la duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare sí cambia los sábados cuando las condiciones climáticas son malas. El mal tiempo afecta la duración de estos viajes específicos en sábado.
